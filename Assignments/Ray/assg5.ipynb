## 2. Mini Cosmic Skirmish — **Simplified Battle-Loop Challenge**

### Description

Two space adventurers, **Nova** and **Zorg**, duel in a tiny arena.
Each turn a fighter must choose **one of just two actions**, implemented as stand-alone functions that you’ll call over and over in a loop:

| Action Function          | Effect                                                        |
| ------------------------ | ------------------------------------------------------------- |
| `attack(caster, target)` | Subtract **ATK** points from *target*’s HP.                   |
| `heal(caster)`           | Add **H** points to *caster*’s HP (but never above `max_hp`). |

Initial combatants:

| Fighter         | `max_hp` | `atk` | `heal_amount (H)` |
| --------------- | -------- | ----- | ----------------- |
| **Nova** (you)  | 100      | 18    | 12                |
| **Zorg** (boss) | 130      | 15    | 10                |

**Goal:** Simulate the duel, print a short log for every turn, and finally announce the winner.

### Requirements

1. **Data Model**
   Either

   * a minimal `Fighter` class **or**
   * two plain dictionaries (one per fighter)
     holding: `name`, `hp`, `max_hp`, `atk`, `heal_amount`.

2. **Action Functions** *(reused every turn)*

   ```python
   def attack(caster, target): ...
   def heal(caster): ...
   ```

   *Both mutate the passed-in objects directly — no return values.*

3. **Turn Logic (`take_turn`)**

   * If a fighter’s HP is **≤ 35 %** of `max_hp`, call **`heal`**.
   * Otherwise call **`attack`**.

4. **Battle Loop**

   ```python
   turn = 1
   while nova["hp"] > 0 and zorg["hp"] > 0:
       take_turn(nova, zorg)      # Nova’s move

       take_turn(zorg, nova)      # Zorg’s move
       turn += 1
   ```

5. **Output Format**

   * Print one line per action, e.g.

